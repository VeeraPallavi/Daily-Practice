```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'day' : [1, 2, 3, 4, 5, 6, 7],
    'sales' : np.random.randint(100, 500, size = 7).tolist()
}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
#Line chart
plt.plot(df["day"], df["sales"], color = "red")
plt.title("Line Plot")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```

```python
data = {
    'name' : ["Pallavi", "Jhon", "Tim", "Lee", "Joe"],
    'age' : [23, np.nan, 45, 33, np.nan]
}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
#Fill missing ages with the mean.
mean = np.round(df['age'].mean(), 2)
df.fillna(mean, inplace = True)
df
```

```python
#Plot a bar chart of name vs age.
plt.bar(df['name'], df['age'], color = 'blue')
plt.title("name vs age")
plt.xlabel("Name")
plt.ylabel("Age")
plt.show()
```

```python
data = {
    'department' : ["HR", "IT", "Finance", "HR", "IT", "HR", "Finance"],
    'salary' : [25000, 30000, 35000, 28000, 32000, 25000, 30000]
}
df = pd.DataFrame(data)
df
```

```python
#Calculate average salary per department.
grouped_data = df.groupby('department')['salary'].mean()
grouped_data
```

```python
#Visualize using a bar chart.
plt.bar(df['department'], df['salary'], color = "Yellow")
plt.title("Bar Chart")
plt.xlabel("Department")
plt.ylabel("salary")
plt.show()
```

```python
gender = ['M', 'F', 'F', 'M', 'M', 'F', 'M', 'F', 'F', 'F']

df = pd.DataFrame(gender)

count = pd.value_counts(df[0])
count
```

```python
data = {
    'month' : ['Jan', 'Feb', 'Mar','April', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'revenue' : [20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000, 70000, 75000],
    'expenses' : [10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000]
}
df = pd.DataFrame(data)
df
```

```python
data = {
    'day' : [1, 2, 3, 4, 5, 6, 7],
    'daily_sales' : [10, 20, 30, 30, 40, 50, 60]
}
df = pd.DataFrame(data)
df
```

```python
#Compute cumulative sales.
cumlative_sales = df['daily_sales'].cumsum()
cumlative_sales
```

```python
#Plot cumulative sales using a line chart.
plt.plot(cumlative_sales, color = "green")
plt.title("cumulative sales")
plt.xlabel("Days")
plt.ylabel("Cumulative sales")
plt.show()
```

```python
data = {
    "product_category": ["Electronics", "Clothing", "Electronics", "Groceries", "Clothing", "Electronics", "Groceries", "Clothing"]
}

df = pd.DataFrame(data)
df
```

```python
#Count occurrences.
category_counts = df["product_category"].value_counts()
category_counts
```

```python
#Plot a bar chart showing product frequency.
plt.bar(category_counts.index, category_counts.values)
plt.title("Product Category Frequency")
plt.xlabel("Product Category")
plt.ylabel("Frequency")
plt.show()
```

```python
data = {
    "city": ["Mumbai", "Delhi", "Bengaluru", "Chennai", "Hyderabad"],
    "population": [20400000, 16700000, 12400000, 11000000, 10000000]
}

df = pd.DataFrame(data)
df
```

```python
#Filter cities with population greater than 1 million.
filtered_data = df[df['population'] > 10000000]
filtered_data
```

```python
plt.bar(filtered_data['city'], filtered_data['population'], color = "Orange")
plt.title("Bar chart for the filtered cities")
plt.xlabel("city")
plt.ylabel("Population")
plt.show()
```
